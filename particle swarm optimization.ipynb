```python
import numpy as np
w = 0
c1 = 0.2
c2 = 0.5
```

```python
xt = np.random.normal(size=5)
vt = np.random.normal(size=5)
```

```python
def fitness(x):
    return 1+2*x-x**2
```

```python
def update(xt,vt,pbest,gbest):
    return w*vt+c1*np.random.uniform(size=1)*(pbest-xt)+c2*np.random.uniform(size=1)*(gbest-xt)    
```

```python
y = fitness(xt)
```

```python
pbest = np.zeros(5)
gbest = np.zeros(1)

for iteration in range(1000):
    y = fitness(xt)
    print('y',y)
    print('xt',xt)
    if iteration == 0:
        pbest = xt
    else:
        pbest = (fitness(pbest)<y)*xt + (fitness(pbest)>=y)*pbest        
    gbest = (fitness(gbest)>y.max())*gbest +(fitness(gbest)<=y.max())*xt[y.argmax()]
    print('gbest',gbest, fitness(gbest))
    print('pbest', pbest)
    xt += update(xt,vt,pbest,gbest)
```
